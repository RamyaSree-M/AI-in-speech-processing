```python
import librosa
import matplotlib.pyplot as plt

# Load the audio file
y, sr = librosa.load(r"C:\Users\himas\Downloads\sem6_learning\speechProcessing\ai_in_speech.mp3")

# Display the original speech signal
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Original Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()


# Trim the silent parts from the beginning and end of the signal
y_trimmed,index= librosa.effects.trim(y)

# Display the trimmed speech signal
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_trimmed, sr=sr)
plt.title('Trimmed Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
#yt, index = librosa.effects.trim(y)
print(librosa.get_duration(y=y), librosa.get_duration(y=y_trimmed))
```

```python
# A2. Use librosa.effects.split() to split the recorded speech with detected silences.
# Play around with the top_db parameter and see the effects of split. 
# Listen to the generated signals and observe the split quality
```

```python
import librosa
import sounddevice as sd

# Load the audio file
y, sr = librosa.load(r"C:\Users\himas\Downloads\sem6_learning\speechProcessing\lab3\voice_lab1.mp3")

# Split the speech signal into segments based on detected silences
segments = librosa.effects.split(y, top_db=20)  # Adjust top_db as needed

# Listen to each segment
for i, segment in enumerate(segments):
    start, end = segment
    segment_audio = y[start:end]

    print(f"Playing Segment {i+1}: Duration {len(segment_audio)/sr} seconds")
    sd.play(segment_audio, sr)
    sd.wait()
```
